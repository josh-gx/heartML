## Overview

**Context**

This database contains 76 attributes, but all published experiments refer to using a subset of 14 of them. In particular, the Cleveland database is the only one that has been used by ML researchers to this date. 

**Variables**

1. age  = in years
2. sex  = (1 = male; 0 = female)
3. cp  = chest pain type (4 values) 
4. trestbps  = resting blood pressure in mm Hg
5. chol  = serum cholestoral in mg/dl 
6. fbs  = fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
7. restecg  = resting electrocardiographic results (values 0,1,2)
8. thalach  = maximum heart rate achieved 
9. exang  = exercise induced angina (1 = yes; 0 = no)
10. oldpeak  = ST depression induced by exercise relative to rest 
11. slope  = the slope of the peak exercise ST segment 
12. ca  = number of major vessels (0-3) colored by flourosopy 
13. thal  = thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
14. **target  = presence of heart disease (1 = true; 0 = false)**

**Creators** 
1. Hungarian Institute of Cardiology. Budapest: Andras Janosi, M.D. 
2. University Hospital, Zurich, Switzerland: William Steinbrunn, M.D. 
3. University Hospital, Basel, Switzerland: Matthias Pfisterer, M.D. 
4. V.A. Medical Center, Long Beach and Cleveland Clinic Foundation: Robert Detrano, M.D., Ph.D.

Data Source: https://www.kaggle.com/ronitf/heart-disease-uci

**Objecive:** Classify patients for presence of heart disease

# Model Data Preparation

## Encode Categorical Variables

```python
# categorical variables reference
# binary: sex, fbs  exang, target
print('categorical variables:', vars_cat, 
      '\n total: ', len(vars_cat))
```

```python
# encode categorical variables for model (5 non binary)
# cp, restecg, slope, ca, thal 

a = pd.get_dummies(df['cp'], prefix = 'cp')
b = pd.get_dummies(df['restecg'], prefix = 'restecg')
c = pd.get_dummies(df['slope'], prefix = 'slope')
d = pd.get_dummies(df['ca'], prefix = 'ca')
e = pd.get_dummies(df['thal'], prefix = 'thal')

dummies = [df, a, b, c, d, e]

# add new variables
model_df = pd.concat(dummies, axis = 1) 

# drop non-encoded variables
model_df = model_df.drop(columns = ['cp', 'restecg', 'slope', 'ca', 'thal']) 
```

```python
# quick data model check on new variables
print(list(model_df.columns))
print(' \n Total Variables: ', len(model_df.columns))
```

## Train & Test Data Split

```python
# split by features and target

from sklearn.model_selection import train_test_split

X = model_df.loc[:, model_df.columns != 'target'] # all except y

y = model_df['target']
```

```python
# train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Machine Learning

```python
# Evaluaton Metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_curve, auc
import sklearn.metrics as metrics
```

## Logistic Reggression

```python
# fit logistic reggression model

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train,y_train)

clf_predictions = clf.predict(X_test)
```

### Classification Report

```python
print('Logistic Regression Scores: \n', classification_report(y_test,clf_predictions))
```

### Confusion Matrix

```python
# Confusion Matrix df
clf_cm = confusion_matrix(y_test, clf_predictions)

clf_cm_df = pd.DataFrame(clf_cm,
                     index = ['Active = 0 ','Active = 1'], 
                     columns = ['Active = 0','Active = 1'])
```

```python
# Confusion Matrix Heatmap
plt.figure(figsize=(10,4))
sns.heatmap(clf_cm_df, annot=True)
plt.title('Logistic Regression \n Accuracy: {0:.3f}'.format(accuracy_score(y_test, clf_predictions)))
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

### ROC & AUC

```python
# ROC & AUC

clf_probs = clf.predict_proba(X_test)
clf_preds = clf_probs[:,1]

fpr, tpr, threshold = metrics.roc_curve(y_test, clf_preds)
clf_roc_auc = metrics.auc(fpr, tpr)
```

```python
# ROC Plot
plt.title('Logistic Regression ROC')
plt.plot(fpr, tpr, 'c', label = 'AUC = %0.2f' % clf_roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

## Random Forest

```python
# fit random forest 

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train,y_train)

rfc_predictions = rfc.predict(X_test) 
```

### Classification Report

```python
# random forest scores

print('Random Forest: \n', classification_report(y_test,rfc_predictions))
```

### Confusion Matrix

```python
# Confusion Matrix df
rfc_cm = confusion_matrix(y_test, rfc_predictions)

rfc_cm_df = pd.DataFrame(rfc_cm,
                     index = ['Active = 0 ','Active = 1'], 
                     columns = ['Active = 0','Active = 1'])
```

```python
# Confusion Matrix Heatmap
plt.figure(figsize=(10,4))
sns.heatmap(rfc_cm_df, annot=True)
plt.title('Random Forest \n Accuracy: {0:.3f}'.format(accuracy_score(y_test, rfc_predictions)))
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

### ROC & AUC

```python
rfc_probs = rfc.predict_proba(X_test)
rfc_preds = rfc_probs[:,1]

fpr, tpr, threshold = metrics.roc_curve(y_test, rfc_preds)
roc_auc = metrics.auc(fpr, tpr)
```

```python
plt.title('Random Forest ROC')
plt.plot(fpr, tpr, 'c', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

#  Results Summary

```python
# logistic regression metrics summary 
clf_accuracy = metrics.accuracy_score(y_test, clf_predictions)
clf_precision = metrics.precision_score(y_test, clf_predictions)   
clf_recall =  metrics.recall_score(y_test, clf_predictions)
clf_f1 = metrics.f1_score(y_test, clf_predictions)
clf_auc = metrics.roc_auc_score(y_test, clf_predictions)                                                                        
```

```python
# logistic regression scores

clf_results = {'Accuracy': [clf_accuracy], 
               'Precision': [clf_precision],
               'Recall': [clf_recall],
               'F1': [clf_f1],
               'AUC': [clf_auc]}
               
clf_results_df = pd.DataFrame(clf_results)
```

```python
# Random Forest metrics summary 
rfc_accuracy = metrics.accuracy_score(y_test, rfc_predictions)
rfc_precision = metrics.precision_score(y_test, rfc_predictions)   
rfc_recall =  metrics.recall_score(y_test, rfc_predictions)
rfc_f1 = metrics.f1_score(y_test, rfc_predictions)
rfc_auc = metrics.roc_auc_score(y_test, rfc_predictions)                                                                        
```

```python
# Random Forest scores , orient ='h'

rfc_results = {'Accuracy': [rfc_accuracy], 
               'Precision': [rfc_precision],
               'Recall': [rfc_recall],
               'F1': [rfc_f1],            
               'AUC': [rfc_auc]}
               
rfc_results_df = pd.DataFrame(rfc_results)
```

```python
clf_results_df['Model'] = 'LogReg'
rfc_results_df['Model'] ='RanForest'
```

```python
all_results = pd.concat([clf_results_df,rfc_results_df])
all_results.set_index('Model', inplace=True )
```

## 70/30 Split Performance Summary

```python
plt.figure(figsize=(14,2.5))
sns.heatmap(all_results.round(2), annot=True)
plt.title('Total Performance')
plt.ylabel('')
plt.xlabel('Scores')
```

## K-Fold Performance

```python
from sklearn.cross_validation import cross_val_score, cross_val_predict

# run models
k_folds = 10
clf_kfold_scores = cross_val_score(clf, X, y, cv=k_folds)
rfc_kfold_scores = cross_val_score(rfc, X, y, cv=k_folds)

# results
s1 = pd.Series(clf_kfold_scores, name='LogReg')
s2 = pd.Series(rfc_kfold_scores, name='RanForest')
```

```python
plt.figure(figsize=(14,5))

ax = plt.axes()
x = np.linspace(0, 10)

ax.plot(s1)
ax.plot(s2)

plt.title('Cross Validation Performance')
plt.ylabel('Scores')
plt.xlabel('K-Folds')
plt.legend()
```
